```python
from datetime import datetime as dt 
import pandas as pd 
import json
import seaborn as sns
import matplotlib.pyplot as plt

today = dt.today().date()
filename = input("Enter your filename: ").strip().lower()
if not filename.endswith(".json"):
    filename += ".json"

listed = []


def save_list():
    with open(filename, "w") as f:
        json.dump(listed, f, indent = 4)
        print("List Saved Successfully for today")

def open_list():
    global listed
    try:
        with open(filename, "r") as f:
            listed= json.load(f)
            print("Loaded Successfully")
    except FileNotFoundError:
        listed = []
        print("No saved activties found")

def add_list():
    while True:
        print("1. To Add List")
        print("2. To quit")
        
        try:
            to_open = int(input("Enter 1/2"))
            
            if to_open == 2:
                print("You have ended the adding of activities")
                break 
            
            elif to_open == 1:
                automated_date = today
                add_activity = input("Enter the activity").strip().upper()
                listed.append({"date": automated_date.strftime("%d-%m-%Y"),"activity":add_activity, "status":"Not Done"})
                save_list()
            else:
                print("Enter either (1/2)")
        except ValueError:
            print("Invalid input")

def open_activity():
    if not listed:
        print("No activities found yet!")
        return

    print("1. View Today Activities")
    print("2. View other Activities")
    try:
        choice_open = int(input("Enter number to view"))
        if choice_open == 1:
            today_list = today.strftime("%d-%m-%Y")
            today_scrath = [y for y in listed if today_list == y['date']]
            if not today_scrath:
                print("No Activity Found")
                return 
            for index, today_op in enumerate(today_scrath, start =1):
                print(f"{index} | {today_op['activity']} | {today_op['status']}")
        elif choice_open == 2:
            date_to = input("Enter the date to view (dd-mm-yyyy)").strip()
            date_to_open = [t for t in listed if date_to == t['date']]
            if not date_to_open:
                print ("No Acitvity Found With Such Date")
                return
            
            for index, dated_now in enumerate(date_to_open, start=1):
                print(f"{index} | {dated_now['activity']} | {dated_now['status']}")
        else:
            print("Enter (1/2)")
    except ValueError:
        print("Invalid Choice")

def search_activity():
    if not listed:
        print("No activities found yet!")
        return

    while True:
        print("1. To search for an activity")
        print("2. To search for a particular day")
        print("3. To Quit Searching")

        try:
            to_search = int(input("Enter (1/2/3)"))
            if to_search == 3:
                print("Searching Ended")
                break
            elif to_search == 1:
                activi_to_search = input("Enter the activity to search").strip().lower()
                found = False
                for index, searche in enumerate(listed, start=1):
                    if activi_to_search in searche["activity"].lower():
                        print(f"{index} | {searche['activity']} | {searche['date']}")
                        found = True
                if not found:
                    print("Activity entered not found")
                    continue
            
            elif to_search == 2:
                date_to_search = input("Enter the date to search (dd-mm-yyyy)")
                found = False
                for index, datee in enumerate(listed, start=1):
                    if date_to_search in datee["date"]:
                        print(f"{index} | {datee['activity']} | {datee['date']}")
                        found = True
                if not found:
                    print("Date entered not found")
                    continue
            else:
                print("Enter (1/2/3)")
        except ValueError:
            print("Invalid Input")

def tick_activities_done():
    if not listed:
        print("No activities found yet!")
        return

    while True:
        print("1. To tick for today's activities")
        print("2. To quit ")
        try:
            choice = int(input("Enter (1/2)"))
            if choice not in (1, 2 ):
                print("Invalid Choice")
                continue
            if choice == 2:
                print("Quit Successfully")
                break
            
            elif choice == 1:
                ticked = today.strftime("%d-%m-%Y")
                done_time = [doned for doned in listed if ticked == doned["date"]]

                if not done_time:
                    print("No Activities Found for today")
                    continue

                for index, tickeds in enumerate(done_time, start=1):
                    print(f"{index} | {tickeds['activity']} | {tickeds['status']}")
                
                while True:
                    print("1. Start ticking/Unticking ")
                    print("2.  End Ticking")
                    try:
                        to_begin = int(input("Enter (1/2)"))
                        if to_begin == 2:
                            print("Ticking Quitted")
                            break
                        elif to_begin == 1:
                            try:
                                choice_to_tick = int(input("Enter the number for the activity to tick"))
                                done_time_tick = done_time[choice_to_tick -1]

                                if done_time_tick["status"] == "Not Done":
                                    done_time_tick["status"] = "Done"
                                    print(f"Activity '{done_time_tick['activity']}' marked as done ")
                                else: 
                                    done_time_tick["status"] = "Not Done"
                                    print(f"Activity '{done_time_tick['activity']}' marked as not done")
                                
                                save_list()
                            except (ValueError, IndexError):
                                print("Invalid Input")
                        else:
                            print("Enter (1/2)")
                    except ValueError:
                        print("Invalid Input")
        except ValueError:
            print("Invalid Choice")


def delete_list():
    if not listed:
        print("No activities found yet!")
        return

    while True:
        print("1. Delete Activity in Today List")
        print("2. Delete Activity in Past list")
        print("3. Quit Deleting")

        try:
            choice_del = int(input("Enter the number (1/2/3)"))
            if choice_del not in (1, 2, 3):
                print("Enter (1/2/3)")
                continue
            if choice_del == 3:
                print("Deleting Ended")
                break

            elif choice_del == 1:
                ticked_del = today.strftime("%d-%m-%Y")
                save_del = [e for e in listed if ticked_del == e['date']]

                if not save_del:
                    print("No Activity found")
                    continue

                for index, ticked_to in enumerate(save_del, start=1):
                    print(f"{index} | {ticked_to['activity']} | {ticked_to['status']}")

                try:    
                    choice_to_del = int(input("Enter the number to delete")) 
                    if choice_to_del < 1 or choice_to_del > len(save_del):
                        print("Invalid Number")
                        continue

                    deleting = save_del[choice_to_del -1]
                    
                    listed.remove(deleting)
                    save_list()
                except ValueError:
                    print("Invalid Input")
            elif choice_del == 2:
                to_del = input("Enter the date to Find (dd-mm-yyyy)").strip()
                save_to_del = [f for f in listed if to_del == f['date']]

                if not save_to_del:
                    print("No Activity Found")
                    continue

                for index, now_del in enumerate(save_to_del, start=1):
                    print(f"{index} | {now_del['activity']} | {now_del['status']}")

                try:
                    choice_now_del = int(input("Enter the number to delete"))
                    if choice_now_del < 1 or choice_now_del > len(save_to_del):
                        print("Invalid Choice")
                        continue
                    
                    deleting_old = save_to_del[choice_now_del -1 ]
                    listed.remove(deleting_old)
                    save_list()
                except ValueError:
                    print("Invalid Input")
        except ValueError:
            print("Invalid Input")

def edit_list():
    if not listed:
        print("No activities found yet!")
        return

    while True:
        print("1. Edit activity")
        print("2. Quit Editing")
        try:
            choice_edit = int(input("Enter (1/2)"))
            if choice_edit not in (1,2):
                print("No valid Options")
                continue
            
            if choice_edit == 2:
                print("Editting Quited Successfully")
                break 
            elif choice_edit == 1:
                print("1. To edit today's activity")
                print("2. To edit Past Activities")
                try:
                    choice_again = int(input("Enter (1/2)"))
                    if choice_again not in (1, 2):
                        print("No Valid Option")
                        continue
                    if choice_again == 1:
                        edit_today = today.strftime("%d-%m-%Y")
                        list_edit_today = [w for w in listed if edit_today == w['date']]

                        if not list_edit_today:
                            print("No Activity Found")
                            continue
                        
                        for index, editee in enumerate(list_edit_today, start =1):
                            print(f"{index} | {editee['activity']} | {editee['status']}")

                        
                        try:
                            choice_to_edit = int(input("Enter the number to edit"))
                            if choice_to_edit < 1 or choice_to_edit > len(list_edit_today):
                                print("Wrong Input")
                                continue
                        except ValueError:
                            print("Invalid Input")
                            continue
                            
                        today_edit = list_edit_today[choice_to_edit -1]

                        new_activity = input("Enter the new activity: ").strip().upper()
                        if new_activity:
                            today_edit['activity'] = new_activity
                        save_list()
                    elif choice_again == 2:
                        date_edit = input("Enter the date to edit (dd-mm-yyyy): ").strip()
                        date_edit_found = [q for q in listed if date_edit == q['date']]

                        if not date_edit_found:
                            print("No activity found")
                            continue

                        for index, editq in enumerate(date_edit_found, start=1):
                            print(f"{index} | {editq['activity']} | {editq['status']}")

                        try:
                            date_to = int(input("Enter the number to edit"))
                            if date_to < 1 or date_to > len(date_edit_found):
                                print("Wrong Choice")
                                continue
                        except ValueError:
                            print("Invalid Input")
                        
                        date_now_edit = date_edit_found[date_to - 1]

                        new_date_act = input('Enter the Activity').strip().upper()
                        if new_date_act:
                            date_now_edit['activity'] = new_date_act
                            save_list()
                except ValueError:
                    print("Invalid Input")
        except ValueError:
            print("Invalid Input")

def visualize_activity():
    if not listed:
        print("No activities found yet!")
        return

    while True:
        print("1. To Visualize MY Activity Level")
        print("2. Quit Visualization")
        try:
            choice_to_open =int(input("Enter(1/2)"))
            if choice_to_open not in (1,2):
                print("Wrong Input")
                continue
        except ValueError:
            print("Invalid Input")
            continue
        if choice_to_open == 2:
            print("Quiting Visualization ...........")
            break
        elif choice_to_open == 1:
            listed_df = pd.DataFrame(listed)

            if "date" in listed_df.columns:
                listed_df['date'] = pd.to_datetime(listed_df['date'], errors="coerce")

            print("Enter the range of date to visualize")
            start_date = input("Enter the start date seperated by comma (dd-mm-yyyy): ").strip()
            end_date = input('Enter the end date seperated by commas (dd-mm-yyyy): ').strip()

            start_time = pd.to_datetime(start_date, format="%d-%m-%Y", errors="coerce")
            end_time = pd.to_datetime(end_date, format="%d-%m-%Y", errors="coerce")

            if pd.isna(start_time) or pd.isna(end_time):
                print("Invalid, use(dd-mm-yyyy)")
                continue

            range_time = (listed_df['date'] >= start_time) & (listed_df['date'] <= end_time)

            listed_df_chosed= listed_df.loc[range_time]

            if listed_df_chosed.empty:
                print("No activity found")
                continue
            
            listed_df_chosed = listed_df_chosed.sort_values("date")
            g= sns.catplot(x="date", hue='status', data=listed_df_chosed, kind="count")
            plt.xticks(rotation=90)
            title_name = input("Enter your title_name: ")
            g.figure.suptitle(title_name, y=1.05)
            plt.show()
        else:
            print("Enter (1/2)")

print("-------------------------------------------")
print("|               TO-DO LIST APP            |")
print("-------------------------------------------")
print()
print("-------------------")
print("|     WELCOME     |")
print("-------------------")
print()
open_list()

while True:
    print("1. Add TO-DO list") 
    print("2. Open TO-DO list")
    print("3. Search TO-DO list")
    print("4. Tick Done in TO-DO list")
    print("5. Delete TO-DO list")
    print("6. Edit TO-DO list")
    print("7. View Levels in TO-DO list")
    print("8. Quit App")

    try:
        to_start = int(input("Enter the number: "))
        if to_start not in range(1,9):
            print("Wrong Choice")
            continue
        if to_start == 8:
            print("Quit Successfully \nThanks for using our program")
            break
        elif to_start == 1:
            add_list()
        elif to_start == 2:
            open_activity()
        elif to_start == 3:
            search_activity()
        elif to_start == 4:
            tick_activities_done()
        elif to_start == 5:
            delete_list()
        elif to_start == 6:
            edit_list()
        elif to_start == 7:
            visualize_activity()
        else:
            print("Enter from (1-8)")
    except ValueError:
        print("Invalid Input")
    

            



                












                        







                            
                                    


                                



                




```
